```python
import torch

x = torch.tensor([1.0, 2.0, 3.0])
print(x)
```

```python
import torch 
import torch.nn as nn 

class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(2,4)
        self.layer2 = nn.Linear(4,1)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = self.layer2(x)
        return x
```

```python
model = NeuralNetwork()
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
losses = []
for epoch in range(100):
    X = torch.tensor([[1.0, 2.0]])
    y = torch.tensor([[1.0]])

    output = model(X)
    loss = loss_fn(output, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    losses.append(loss.item())

print(losses)
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Curve")
plt.show()
```

```python
import torch 
import torch.nn as nn 

# Data
X = torch.tensor([
    [1.0, 2.0],
    [2.0, 1.0],
    [3.0, 3.0],
    [4.0, 5.0]
])

y = torch.tensor([
    [0.0],
    [0.0],
    [1.0],
    [1.0]
])
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(2,4)
        self.layer2 = nn.Linear(4,1)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = torch.sigmoid(self.layer2(x))
        return x


model = NeuralNetwork()
```

```python
loss_fn = nn.BCELoss() # Used for binary classfication - loss function 
```

```python
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

```python
losses = []

for epoch in range(200):
    output = model(X)
    loss = loss_fn(output, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    losses.append(loss.item())

print("Training done")
```

```python
with torch.no_grad():
    predictions = model(X)
    predicted_classes = (predictions > 0.5).float()
    print(predicted_classes)
```

```python
correct = (predicted_classes == y).sum().item()
accuracy = correct / len(y)

print("Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt

plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.show()
```
